### 개요

캐릭터가 죽지 않기 위해서는 점프와 슬라이딩을 이용해 적절한 타이밍에 장애물을 회피하고, 높은 점수를 받기 위해서는 젤리를 먹어야 합니다.

이 과정은 딥러닝 인공신경망을 통해 학습됩니다. 

여기서

* 캐릭터 : Agent
* 게임화면 : Enviorment

의 역할을 합니다.

![대체 텍스트](https://image.slidesharecdn.com/deviewalpharun-161025082950/95/ai-deview-2016-37-638.jpg?cb=1479942068)

여기서 게임 화면 이미지를 위와 같이 분할하여 각 칸의 상태를 State, S(t) 로 표현합니다.

![대체 텍스트](https://image.slidesharecdn.com/deviewalpharun-161025082950/95/ai-deview-2016-42-638.jpg?cb=1479942068)

이 State 는 Agent가 받아 인공신경망(Neural network)에 입력값으로 들어갑니다. 이 때, 인공신경망을 거쳐 나오는 결과는

* 가만히
* 점프
* 슬라이딩

이 있습니다.

이 결과에 따라 Agent 는 Action 결정 하고 그 행동에 따라 Enviorment 가 영향을 받습니다.

이렇게 Agent 와 Enviorment 는 서로에게 영향을 주는 순환관계를 가지고 있습니다.

이 과정에서 쿠기가 젤리를 먹거나 장애물에 부딪히는 등의 이벤트는 Reward (보상) 으로 쌓이게 되고 이 보상을 기반으로 인공신경망을 학습시킵니다.

즉, 강화학습은
* Agent 가 action 을 결정하는 방법을 학습시키는 것
* 각 action 은 그 다음 state 에 영향을 끼친다.
* 성공한 정도는 reward 로 측정
* 목표 : 미래의 reward 가 최대가 되도록 action 을 취하는 것.

이라고 할 수 있습니다.

# 2. 언어

  NLP(Natural Languuage Processing) 기술은 인공지능의 핵심 기능 중 하나입니다. 1950년대부터 기계 번역과 같은 자연어 처리 기술이 연구되기 시작하여 최근에는 딥러닝을 이용한 언어처리 연구 개발이 전 세계적으로 활발히 진행되고 있습니다.  
  
  저는 언어 처리 기술 중에서도 평소에도 자주 쓰는 번역 기술에 대해 조사했습니다.

-------------------------------------------

사례 : 파파고

> 출처 _ [Papago 소개 | naver developers](https://developers.naver.com/products/nmt/)

현제 네이버 Papago 번역 에는 NMT(Neural Machine Transalation) , 인공신경망 기반 기계번역 이 사용되었습니다.

 NMT 기술은 입력 문장을 문장벡터로 변환하는 신경망(encoder)과 문장벡터에서 번역하는 언어의 문장을 생성하는 신경망(decoder)를 대규모의 병렬 코퍼스부터 자동으로 학습합니다. 

입력문장의 일부가 아니라 문장 전체 정보를 바탕으로 번역을 수행하기때문에 기존 SMT(통계 기반 번역) 방식의 번역보다 더욱 정확하고 문장 맥락에 맞는 번역을 할 수 있습니다.

현제 파파고는 웹 서비스로 사용할 수 있지만, RESTful 형태로 지원되는 API 가 있어, 개발자가 자신의 서비스에 적용할 수도 있습니다.

###기술

번역에 사용되는 기술은
1. 규칙기반번역(RBMT)
2. 통계기반번역(SMT)
3. 인공신경망 기반 기계번역(NMT)

의 순으로 발전되었습니다.

현제 사용되는 nmt 기술에서 encoder 와 decoder 는 각각 RNN(recurrent neural network) 로 구현됩니다.

###RNN 이란

RNN 은 Recurrent Neural Network 의 약자로서 회귀적 특성을 가진 인공 신경망입니다. RNN은 sequence data, 각 요소간 순서가 중요한 연속적인 데이터를 다뤄 딥러닝을 해야할때 일반적인 인공신경망에 비해 효율성을 얻을 수 있습니다.

sequence 데이터를 rnn 네트워크 구조로 처리하면 데이터의 길이에 관계없이 인풋과 아웃풋을 받아들일 수 있으며, 다른 위치에 있는 요소끼리도 학습을 공유할 수 있다는 장점이 있습니다.

언어에서의 문장 또한 시퀀스라고 할 수 있기에 RNN 은 자연어 처리, 번역기술에 사용될 수 있습니다.

카카오 미니의 학습에는 딥러닝(deep learning) 이 사용됩니다.

카카오미니의 두뇌 역할을 하는 카카오I(아이) 의 코어 엔진은 스피커 안에 있는 것이 아닙니다.

카카오미니에는 듣는 역할을 하는 '마이크' 와 사용자에게 말을 하는 역할의 '스피커'가 탑재되어 있고, 사용자가 말한 명령은 인터넷망을 통해 카카오i 코어 엔진 서버로 전송됩니다.

![카카오 AI 생태계](http://t1.daumcdn.net/thumb/R1280x0/?fname=http://t1.daumcdn.net/brunch/service/user/1oU7/image/-YiSidbBctZfzFDr6g4L9SQC0EQ.png)

여기서 입력되는 음성이 서버로 전송되면 다음과 같은 과정을 거칩니다.

1. 음성 엔진 : 음성을 텍스트로 변환
2. 대화 엔진 : 텍스트에서 사용자가 어떤 '의도' 로 말을 했는지 분류하고, 각각의 의미 있는 단어를 '추출'
3. 서비스 에이전트 : 추출된 단어를 이용하여 사용자 '의도' 에 맞는 적절한 답변 생성
4. 스피커 : 생성된 답변을 받아 출력

------------------------------------------
### 기술

카카오 미니를 많이 사용할수록 음성 인식 성능은 점점 좋아집니다. 

그 학습 방법은 아래와 같습니다.

![음성 인식 시스템 학습 과정](http://t1.daumcdn.net/thumb/R1280x0/?fname=http://t1.daumcdn.net/brunch/service/user/1oU7/image/QcrS7da5WC4S_wp-7o9TwnN0vhg.png)

음성 인식 시스템은 크게 '인식' 과정과 '학습' 과정으로 나눌 수 있습니다.

* 인식 과정 : 사용자의 음성을 텍스트 결과로 변환. 이때 음성 엔진은 미리 학습된 음향 모델과 언어 모델을 이용.
* 학습 과정 : 사용자의 음성과 인식된 텍스트를 다시 학습하여 음향 모델과 언어 모델을 만드는 단계. 실제 사용자 데이터는 다양한 연령, 목소리 특성, 주변 환경을 포함한 데이터이므로 이를 학습하여 새로운 모델을 만드는 것은 성능 개선에 필수적인 요소임.

학습 과정에서는 음향 모델과 언어모델이 다시 학습되는데,

여기서 음향 모델은 깊은 신경망(deep neural network, DNN) 방식을 적용하면 대용량의 사용자 데이터를 이용하여 더욱 정교한 모델링이 가능하고, 언어 모델은 GloVe(Global Vectors for Word Representation) 을 활용하여 학습하고 있습니다. 

학습과정을 통해 이용자들이 더 많은 대화를 할수록 스피커는 점점 더 똑똑해집니다.

* 물체 판단 : 딥러닝 이용

  운전자가 주행 도중 전방에 나타난 물체의 정체를 판단하는 일은 매우 중요합니다. 그 물체가 무엇인지에 따라 운전 방식을 바꿔야 하기 때문입니다. 
  사람이 물체를 직관적으로 분류할 수 있는 신경 네트워크를 가지고 있는 것처럼, 딥러닝을 이용해 인간의 학습을 모방하여 물체를 구분합니다.

* 자율주행 자동차를 위한 필수 기능
  1. 인지 :

    카메라, 레이더, 라이다 등 차체 내 센서 정보를 처리해 주변 환경 정보를 알아차림

    * 머신러닝으로 정적 환경정보, 동적 환경 요소를 전부 검출, 분류
  2. 판단 :

    인지된 정보를 이용해 향후 벌어질 일을 예측한 후 가장 안전하고 빠른 차량 궤적을 생성

    * 딥러닝을 이용해 다른 운전자의 움직임을 예측해 차량의 궤적을 생성

  3. 제어:

    최종적으로 생성된 차량 궤적을 정확하게 따라갈 수 있도록 운전대, 액셀러레이터, 브레이크 조작

    * 탑승자의 승차감을 튜닝하기 위해 개개인의 운전 방식을 데이터화해 인간 감성을 고려한 차량 제어

실제로 여러 곳에서 이 세 기능에 딥러닝 기술을 적용하는 연구가 활발하게 진행되고 있습니다.
